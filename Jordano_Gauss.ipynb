```python
""" Калмыков Я.А. КЭ-219
    Метод Жордано - Гаусса """
import numpy as np
import sys






# Исходная матрица
A = [[0, 5, 6,4],
     [7, 8, 9,2],
     [1, 4, 5,3]]

#A = [[0,3,5,6,5],
 #    [0,6,10,12,5],
  #  [4,6,0,2,20],
   # [3,5,7,2,1]]


n = range(len(A))   # Количество строке
m = range(len(A[0]))    # Количество столбцов

leaders = []    # список ведущих элементов в каждой строке
num_str = []    # индексы строк
num_col = []    # индексы столбцов




""" Функция вывода матриц """


def out(C):
    print("Матрица")
    for i in n:
        for j in m:
            print(("{:3.2f}".format(C[i][j])),end='  ')
        print()
    print()

""" Циклы заполнения списков индексов строк и столбцов"""
for i in n:
    num_str.append(i)
for j in m:
    num_col.append(j)


now_str = 0    # Указатель на строку


""" Фукнция для создания нулевой матрицы n*m"""


def zeros(A):
    n = len(A)
    m = len(A[0])
    B = [[0 for j in range(m)] for i in range(n)]
    return B


""" Функция поиска ведущего элемента"""


def lead_el(A):
    i = now_str
    print(" Поиск ведущего элемента в",i+1,"строке")
    for j in m:
        if A[i][j] != 0 and [[i, j]] not in leaders \
                and i in num_str \
                and j in num_col:
            leaders.append([i, j])   # добавление индексов
            # ведущего элемента в общий список
            print('lead is' + str(A[i][j]))
            num_str.remove(i)   # запрет на использование строки
            num_col.remove(j)   # запрет на использование столбца

        print(A[i][j])
    print("\n")


""" Функция пересчёта элементов способом прямоугольника"""


def changer(A, B, now_str):
    lead = leaders[now_str]     # получение индексов ведущего элемента

    i_0 = lead[0]     # i-координата ведущего элемента
    j_0 = lead[1]     # j-координата ведущего элемента
    lead_elem = A[i_0][j_0]   # значение ведущего элемента
    """ Цикл деления строки на ведущий элемент"""
    for c in m:
       B[i_0][c] = A[i_0][c] / lead_elem
    """ Цикл пересчёта всех остальных элементов
        способом прямоугольника"""
    for i in num_str:
        for j in num_col:
            B[i][j] = (A[i][j] * lead_elem) - A[i_0][j] * A[i][j_0]


""" Функция проверки существования решения """

""" Теорема Кронеккера-Капелли — теорема, которая
доказывает: чтобы СЛАУ была совместной, необходимым и достаточным условием
является равенство ранга r матрицы рангу расширенной матрицы.
"""

def check(A):
  leng = len(A)
  ext_m = np.matrix(A)       # расширенная матрица
  init_m = ext_m[:,0:leng]   # исходная матрица 

  # Проверка на равенство ранга исходной и расширенной матриц
  if np.linalg.matrix_rank(ext_m)!= np.linalg.matrix_rank(init_m):
    print(" Ранг исходной матрицы: ",np.linalg.matrix_rank(init_m),'\n'
    ,"Ранг расширенной матрицы: ",np.linalg.matrix_rank(A))
    print(" ",np.linalg.matrix_rank(init_m)," != ",np.linalg.matrix_rank(A)) 
    sys.exit("СЛАУ не совместна!")
  #print(np.linalg.matrix_rank(A))
  #print(np.linalg.matrix_rank(init_m))


""" Основная программа """


check(A)
for i in range(len(A)):
  B = zeros(A)    # создание нулевой матрицы
  lead_el(A)      # поиск ведущего элемента
  changer(A, B, now_str)    # пересчёт всех элементов в матрице
  if now_str == len(A):     # возврат к пропущенной строке
     now_str = 0
  else:
     now_str += 1
  out(A)
  out(B)
  A = B
  """ Циклы заполнения списков индексов строк и столбцов"""
  num_str = []
  num_col = []
  for x in n:
      num_str.append(x)
  for j in m:
      num_col.append(j)

""" Выражение ответа и его вывод в консоль"""
c = len(A[0])
for i in n:
    for j in m:
        if A[i][j] != 0 and j != c-1:
          A[i][c-1] = A[i][c-1]/A[i][j]
          A[i][j] = 1
#print(A)
out(A)

X = []
for i in n:
    X.append(A[i][c-1])
print('Ответ:\n')
i = 1
j = 0
while j < len(X):
    print("x%d=" % i, X[j])
    j += 1
    i += 1

```
